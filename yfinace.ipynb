```python
import yfinance as yf
import datetime as dt
import pandas as pd
```

```python
FacebookInfor = yf.Ticker("META")
print(FacebookInfor.info)

AmazonInfor = yf.Ticker("AMZN")
print(AmazonInfor.info)

GoogleInfor = yf.Ticker("GOOG")
print(GoogleInfor.info)
```

```python
print("Company Sector : ", FacebookInfor.info['sector'])
print("Price Earnings Ratio : ", FacebookInfor.info['trailingPE'])
print("Company Beta : ", FacebookInfor.info['beta'])
```

```python
for key, value in FacebookInfor.info.items():
    print(key, ":", value)
```
